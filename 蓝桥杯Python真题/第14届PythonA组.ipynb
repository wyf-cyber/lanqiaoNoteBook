### 试题 A: 特殊日期（5分）
**问题描述:**
> 记一个日期为 yy 年 mm 月 dd 日，统计从 2000 年 1 月 1 日到 2000000 年 1 月 1 日：有多少个日期满足年份 yy 是月份 mm 的倍数，同时也是 dd 的倍数。

**答案提交:**
> 这是一道结果填空的题，你只需要算出结果后提交即可。本题的结果为一个整数，在提交答案时只输出这个整数，输出多余的内容将无法得分。

**题目答案：**
> 35813063

**要点：**
> 闰年的二月是29天，非闰年的二月是28天

```python
# BF Algorithm
month = [
    [0,31,28,31,30,31,30,31,31,30,31,30,31],
    [0,31,29,31,30,31,30,31,31,30,31,30,31],
]
def isRun(year):
    if year%400 == 0 or (year%4==0 and year%100 != 0):
        # 是闰年
        return 1
    return 0
# loop
ans = 0
for year in range(2000,2000000):
    monthes = month[isRun(year)]
    for m in range(1,13):
        for d in range(1, monthes[m]+1):
            if year%m == 0 and year%d == 0:
                ans += 1
print(ans+1) # 2000000.1.1也是一个解
```

### 试题 B: 分糖果（5分）
**问题描述:**
> 两种糖果分别有9个和16个，要全部分给7个小朋友，每个小朋友得到的糖果总数最少为2个最多为5个，问有多少种不同的分法，糖果必须全部分完。如果有其中一个小朋友在两种方案中分到的糖果不完全相同，这两种方案就算作不同的方案。

**答案提交:**
> 这是一道结果填空的题，你只需要算出结果后提交即可。本题的结果为一个整数，在提交答案时只输出这个整数，输出多余的内容将无法得分。

**题目答案：**
> 35813063

**要点：**
> 注意递归处理的子问题是单个小朋友的分配问题，不是某一颗糖的分配问题，因为糖会出现重复，无法去重

```python
# DFS
ans = 0
def dfs(depth,a,b):
    # depth:为第depth个小朋友分配糖果
    # a:第一种糖果的剩余数量
    # b:第二种糖果的剩余数量
    if depth == 7:
        if a == 0 and b == 0:
            # 找到一种可行解
            ans += 1
        return
    # for循环去重
    for i in range(6):
        for j in range(6):
            # 给当前的小朋友i个A种糖，j个B种糖
            if 2 <= i+j <= 5 and i <= a and j <= b:
                dfs(depth+1,a-i,b-j)
    return
dfs(0,9,16)
print(ans)
```

### 试题 C: 三国游戏（10分）
**问题描述:**
>  小蓝正在玩一款游戏，游戏中魏（X）、蜀（Y）、吴（Z）三个国家各自拥有一定数量的士兵X、Y、Z（一开始可以认为都是0）。游戏有n个可能会发生的事件，每个事件之间相互独立且最多只发送一次，当第i个事件发送时会分别让X、Y、Z增加$A_i,B_i,C_i$ .当游戏结束时（所有事件的发生与否已经确定），如果X,Y,Z的其中一个大于另外两个之和，我们认为其获胜。例如，当X>Y+Z时，我们认为魏国获胜，小蓝想知道游戏结束时，如果有其中一个国家获胜，最多发生了多少个事件？如果不存在任何一个让某个国家获胜的情况，请输出-1。

**输入格式:**
> 输入的第一行包含一个整数n。\
第二行包含n个整数表示A，相邻整数之间使用一个空格分隔\
第三行包含n个整数表示B，相邻整数之间使用一个空格分隔\
第四行包含n个整数表示C，相邻整数之间使用一个空格分隔

**输出格式**
> 输出一行包括一个整数表示答案。

**要点：**
> 注意题目要求游戏结束时胜利的情况下的最大事件数，可以使用贪心算法

```python
n = int(input())
a = [int(i) for i in input().split()]
b = [int(i) for i in input().split()]
c = [int(i) for i in input().split()]
d_x = [a[i]-b[i]-c[i] for i in range(n)]
d_y = [b[i]-a[i]-c[i] for i in range(n)]
d_z = [c[i]-a[i]-b[i] for i in range(n)]
d_x.sort(reverse = True)
d_y.sort(reverse = True)
d_z.sort(reverse = True)
ans = 0
resX,resY,resZ = 0,0,0
sumX,sumY,sumZ = 0,0,0
# 魏国获胜的条件下最多发生的事件数
for i in range(n):
    sumX += d_x[i]
    if sumX <= 0:
        # lose
        resX = i # 0~i-1一共i个事件
        break
for i in range(n):
    sumY += d_y[i]
    if sumY <= 0:
        resY = i
        break
for i in range(n):
    sumZ += d_z[i]
    if sumZ <= 0:
        resZ = i
        break
# 三国获胜的情况不能兼容，所以取最大值
ans = max(resX,resY,resZ)
print(ans)
```

### 试题 D: 平均（10分）
**问题描述:**
>  有一个长度为n的数组(n是10的倍数)，每个数a都是区间[0, 9]中的整数。小明发现数组里每种数出现的次数不太平均，而更改第i个数的代价为bi，他想更改若干个数的值使得这10种数出现的次数相等(都等于n/10)，请问代价和最少为多少。

**输入格式:**
> 输入的第一行包括一个正整数n。\
接下来n行，第i行包括两个整数$a_i$，$b_i$，用一个空格分隔。

**输出格式**
> 输出一行包括一个正整数表示答案。

**要点：**
> 因为已经知道从当前状态到目标状态各个种类的数所需要转换的数量，所以只需要从当前数中选出最小修改代价的数即可。

```python
n = int(input())
pre = n // 10
nums = [[0,0] for _ in range(n)]
count = [0]*10  # 记录各个数字的数量
for i in range(n):
    nums[i][0],nums[i][1] = map(int,input().split())
    count[nums[i][0]] += 1
nums.sort()
ans = 0
for i in range(10):
    if count[i] <= pre:
        continue
    # need change count[i]-pre 
    # find this one
    p = 0
    while nums[p][0] != i:
        p += 1
    for k in range(count[i]-pre):
        # 将多余的数字转化成别的数
        ans += nums[p+k][1]
print(ans)
```

### 试题 E: 翻转（15分）
**问题描述:**
>  小蓝用黑白棋的n个棋子排成了一行，他在脑海里想象出了一个长度为 n的01串T，他发现如果把黑棋当做1，白棋当做0，这一行棋子也是一个长度为n的01串S。小蓝决定，如果在S中发现一个棋子和它两边的棋子都不一样，就可以将其翻转变成另一个颜色。也就是说，如果S中存在子串 101或者010，就可以选择将其分别变为111和000，这样的操作可以无限重复。
小蓝想知道最少翻转多少次可以把S变成和T一模一样

**输入格式:**
> 输入包含多组数据\
输入的第一行包含一个正整数 D表示数据组数\
后面2D行每行包含一个01串，每两行为一组数据，第 2i-1行为第i组数据的Ti，第2i行为第i组数据的Si，Si和Ti度均为n。

**输出格式**
>  对于每组数据，输出一行包含一个整数，表示答案，如果答案不存在请输出-1。

**要点：**
> 注意到在进行一次翻转以后不可能会产生新的可翻转节点，所以不需要使用迭代搜索。翻转后的节点不会再翻转回来，所以不必设置visited列表，使用for循环直接搜索即可。

```python
d = int(input())
for _ in range(d):
    t = [int(i) for i in input()]
    s = [int(i) for i in input()]
    n = len(s)
    if s[0] != t[0] or s[-1] != t[-1]:
        print(-1)
        continue
    count = 0
    for i in range(1,n-1):
        if s[i] == t[i]:
            continue
        else:
            if s[i-1] == s[i+1] and s[i-1] != s[i]:
                count += 1
            else:
                count = -1
                break
    print(count)
```

### 试题 F: 子矩阵（15分）
**问题描述:**
>  给定一个n x m(n行m列)的矩阵。设一个矩阵的价值为其所有数中的最大值和最小值的乘积。求给定矩阵的所有大小为a xb (a 行列)的子矩阵的价值的和。答案可能很大，你只需要输出答案对998244353 取模后的结果。

**输入格式:**
> 输入的第一行包含四个整数分别表示n，m，a，b，相邻整数之间使用一个空格分隔。接下来
n行每行包含m个整数，相邻整数之间使用一个空格分隔，表示矩阵中的每个数$A_{ij}$。

**输出格式**
>  输出一行包含一个整数表示答案。

**要点：**
> 

```python
MOD = 998244353
n,m,a,b = map(int,input().split())
# 原矩阵
matrix = []
for _ in range(n):
    matrix.append([int(i) for i in input().split()])
# 利用优先队列将查找滑动窗口中最值的操作的事件复杂度优化为线性
# 注意输入的nums是一个一维数组，二维的问题需要转换成一维才能解决
def get_max(nums,step):
    # q是按照元素大小存储各元素下标的优先队列，队列中下标对应的元素大小从大到小递减
    q = []
    # max_list存储各个窗口内的最大值
    max_list = []
    for i in range(len(nums)):
        # 从队尾开始弹出队列中所有小于当前下标所对应的元素值的元素下标
        # 因为前面的元素一定比当前元素先弹出，从此时开始区间包含前面元素的时候也一定会包含当前元素
        # 所以区间中的最大元素不可能是小于当前元素的前驱节点，所以可以直接弹出
        while q and nums[q[-1]] < nums[i]:
            q.pop()
        q.append(i)
        # 检查队列长度
        if q[0] <= i - step:
            q.pop(0)
        # 建立有效队列以后，取队头（最大元素）
        if i >= step-1:
            max_list.append(nums[q[0]])
    return max_list
# 同样的原理求出滑动窗口在各个位置的最小元素
def get_min(nums,step):
    q = []
    min_list = []
    for i in range(len(nums)):
        while q and nums[q[-1]] > nums[i]:
            q.pop()
        q.append()
        # 因为i每次递增1，所以每一轮中最多只有一个元素因为离开搜索窗口而需要被弹出
        if q[0] <= i-step:
            q.pop(0)
        if i >= step-1:
            min_list.append(nums[q[0]])
    return min_list

# 先求出行的最大最小值
# 列数被压缩，行数不变
row_max = []
row_min = []
for i in range(n):
    row_max.append(get_max(matrix[i],b))
    row_min.append(get_min(matrix[i],b))
# 将两个矩阵转置后再次用滑动窗口法即可求出每个子矩阵的最值
# 将行数转置成列数，然后压缩列数
row = len(row_max)
col = len(row_max[0])
temp_max = [[0]*row for _ in range(col)]
temp_min = [[0]*row for _ in range(col)]
for i in range(row):
    for j in range(col):
        temp_max[i][j] = row_max[j][i]
        temp_min[i][j] = row_min[j][i]
# 进一步压缩转置后的中间矩阵获得每一个a*b的子矩阵的最大值和最小值所组成的矩阵
sub_max = []
sub_min = []
for i in range(col):
    sub_max.append(get_max(temp_max[i],a))
    sub_min.append(get_min(temp_min[i],a))
for i in range(len(sub_max)):
    for j in range(len(sub_min[0])):
        ans += (sub_max[i][j]*sub_min[i][j])%MOD
        ans = ans%MOD
print(ans)
```

### 试题 G: 阶乘的和（20分）
**问题描述:**
> 给定n个数$A_{i}$，问能满足m!为$\sum_{i=1}^{m}(A_{i}!)$的因数的最大的m是多少，其中m!表示m的阶乘，即1 x 2 x 3 x···x m。

**输入格式:**
> 输入的第一行包含一个整数n。第二行包含n个整数，分别表示$A_{i}$，相邻整数之间使用一个空格分隔。

**输出格式**
> 输出一行包含一个整数表示答案。

**要点：**\
 在特殊情况下，a数组中各个值均只出现一次，各阶乘无法合并，那么m! = a[0]!,a[0] = min(a)\
 本题中各个a[i]可能重复出现，因此存在合并阶乘的可能（比如n!*(n+1) = (n+1)!)，所以m! = min(s)
 s是合并化简后的a数组各元素阶乘和式子中阶乘元素的集合\
 通过以上分析可知m!应当是分子中各个阶乘合并后最小的阶乘\
 通过分治求出合并后的最小阶乘\
 temp = (x\*a[0] + y\*(a[0]+1)),x,y分别是a[0],a[0]+1在a中的个数\
 显然a[0]<=m，现在要看合并结果是否是a[0]+1的倍数，(a[0]+1)*y显然是(a[0]+1)的倍数
 > 倍数性质：m的倍数+不是m倍数的数，则结果一定不是m的倍数

 由倍数定律可知如果x*a[0]是(a[0]+1)的倍数，则temp就是(a[0]+1)的倍数\
 又因为gcd(a[0],a[0]+1) = 1，所以‘x\*a[0]是(a[0]+1)的倍数’等价于x是(a[0]+1)的倍数

```python
from collections import defaultdict
n = int(input())
a = sorted([int(i) for i in input().split()])

# 统计a中各种元素出现的次数
times = defaultdict(int) # 如果没有出现过，就为其赋初值为0
for i in a:
    times[i] += 1

m = a[0]
x = times[m]
# 从最小的元素开始向后递推检查是否可以利用元素的系数增大m的值
while x%(m+1) == 0:
    m += 1
    # 注意此处合并的步长是1，所以原数组中不一定存在这个值，但是由默认字典的特性该项将会自动初始化为0
    # 所以不必做特殊的区分处理
    times[m] += x//m
    x = times[m]
print(m)
```

### 试题 I: 子树的大小（25分）
**问题描述:**
> 给定一棵包含n个结点的完全m又树，结点按从根到叶、从左到右的顺序依次编号。例如下图是一个拥有 11个结点的完全3又树。你需要求出第k个结点的子树的结点数。

**输入格式:**
>  输入包含多组询问。输入的第一行包含一个整数T，表示询问次数。接下来T行，每行包含三个整数 n，m，k，表示一组询问。

**输出格式**
>  输出T行，每行包含一个整数表示对应询问的答案。

**要点：**
> 了解完全m叉树的基本性质，等比数列前n项和

```python
from math import ceil,log
# 求出第k个节点在完全m叉树中对应的子树节点的总个数
T = int(input())
for _ in range(T):
    n,m,k = map(int,input().split())
    depth = ceil(log(n*(m-1)+1, m))
    rootDepth = ceil(log(k*(m-1)+1, m))
    # 相当于求以第k个节点为根节点的子树的节点总数
    # 先求出子树除倒数第一层以外的节点总和
    if rootDepth < depth:
        ans = (m**(depth-rootDepth)-1)//(m-1)
        # 求出k节点在本层中的序号，以便确认子树在最后一层的范围
        indexK = k - (m**(rootDepth-1)-1)//(m-1)
        # 求最后一层节点的个数
        nodes = n - (m**(depth-1)-1)//(m-1)
        if nodes > (indexK-1)*(m**(depth-rootDepth)):
            # 将子树在最后一层的节点添加到答案中
            # 最后一层的子树节点最多不超过 m**(depth-rootDepth)
            ans += min(nodes - (indexK-1)*(m**(depth-rootDepth)), m**(depth-rootDepth))
    else:
        ans = 1
    print(ans)
```

### 试题 J: 反异或01串（25分）
**问题描述:**
> 初始有一个空的01串，每步操作可以将0或1添加在左侧或右侧。也可以对整个串进行反异或操作: 取s' = rev(s)，其中s 是目前的01串，由表示逐位异或，rev(s)代表将 s翻转，也就是说取中心位置并交换所有对称的两个位置的字符。例如，rev(0101)=1010，rev(010) = 010，rev(0011)=1100。\
反异或操作最多使用一次 (可以不用，也可以用一次)。给定一个01串T，问最少需要添加多少个1才能从一个空01 串得到T，在本题中0可以添加任意个。

**输出格式**
> 输出一行包括一个整数，表示最少需要添加多少个1.

**要点：**\

```python
t = [i for i in input()]
# 在列表元素之间插入#确保列表长度为奇数
n = len(t)
 
```
